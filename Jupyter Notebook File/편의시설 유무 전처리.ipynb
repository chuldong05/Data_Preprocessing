### 수유실 시작

```python
Feeding_Room_Df =  pd.read_csv('C:/Users/dlwhd/Downloads/2023 Hackton/Info_Data/(엘베,휠체어리프트,수유 여부)서울시 지하철 역사 편의시설 현황.csv', encoding='cp949')
```

```python
# 문자열 앞의 0 제거
Feeding_Room_Df['호선'] = Feeding_Room_Df['호선'].str.lstrip('0')
Feeding_Room_Df
```

```python
# LINE_NUM 매핑 딕셔너리
line_mapping = {
    '1호선': '1',
    '2호선': '2',
    '3호선': '3',
    '4호선': '4',
    '5호선': '5',
    '6호선': '6',
    '7호선': '7',
    '8호선': '8',
    '9호선': '9',
    # 나머지 매핑값 추가
    '10':'경의선',
    '11':'경춘선',
    '12':'경강선',
    '13':'수인분당',
    '14':'신분당',
    '15':'에버',
    '16':'의정부경전철',
    '17':'인천선',
    '18':'인천2호선',
    '19':'공항철도',
    '20':'신림선',
    '21':'용인경전철',
    '22':'우이신설경전철',
    '23':'서해선',
    '24':'김포도시철도'
}

# 'LINE' 컬럼 추가
Feeding_Room_Df['호선'] = Feeding_Room_Df['호선'].map(line_mapping).fillna('')

Feeding_Room_Df.head(30)
```

```python
Feeding_Room_Df['유아수유방유무'] = Feeding_Room_Df['유아수유방유무'].map({'Y': 1, 'N': 0})
Feeding_Room_Df['유아수유방유무'] = Feeding_Room_Df['유아수유방유무'].astype(str)
Feeding_Room_Df
```

```python
# CSV 파일로 저장
Feeding_Room_Df.to_csv('(최종 전처리) 엘베,휠체어리프트,수유 여부 - 서울시 지하철 역사 편의시설 현황.csv',encoding='cp949',index=False)
```
