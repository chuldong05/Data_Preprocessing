```python
def get_statn(statn):
    a = statn['지하철역ID'].to_list()
    b = statn['이전지하철역ID'].to_list()
    c = statn['다음지하철역ID'].to_list()
    d = statn['지하철호선ID'].to_list()
    e = statn['지하철역명'].to_list()
    for i in range(len(a)):
        a[i] = a[i].replace("\uFEFF", "")
        b[i] = b[i].replace("\uFEFF", "")
        c[i] = c[i].replace("\uFEFF", "")
        d[i] = d[i].replace("\uFEFF", "")
        e[i] = e[i].replace("\uFEFF", "")

    statn['지하철역ID'] = a    
    statn['이전지하철역ID'] = b
    statn['다음지하철역ID'] = c
    statn['지하철호선ID'] = d
    statn['지하철역명'] = e

    return statn

statn = get_statn(statn)
```

#### 1~8호선 까지는 반복문 처리

```python
import re

# EV_Line_1_Df부터 EV_Line_8_Df까지 반복
for i in range(1, 9):
    # EV_Line_1_Df, EV_Line_2_Df, ..., EV_Line_8_Df 변수명 생성
    ev_line_df = globals().get(f"EV_Line_{i}_Df")

    # "역명" 컬럼 중복 제거 및 컬럼 이름 변경
    ev_line_df = ev_line_df.drop_duplicates(subset='역명')
    ev_line_df.rename(columns={'역명': '역명칭'}, inplace=True)
    ev_line_df = ev_line_df[['선명', '역명칭']]

    # "역명칭" 컬럼 값이 일치하는 인덱스 검색
    matching_indices = []
    for index, row in df2.iterrows():
        station_name = row['역명칭']
        pattern = r"\b{}\b".format(station_name.replace("(", r"\(").replace(")", r"\)"))
        regex = re.compile(pattern)
        if any(regex.search(station) for station in ev_line_df['역명칭']):
            matching_indices.append(index)

    # 일치하는 인덱스에 대해 "엘리베이터" 컬럼에 1 표시
    df2.loc[matching_indices, '엘리베이터'] = 1
    
```
