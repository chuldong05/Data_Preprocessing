```python
import requests
import json

base_url = "https://openapi.kric.go.kr/openapi/vulnerableUserInfo/stationSafetyPlatform?serviceKey=$2a$10$vRcwk0uYSEf4P7c0wAlQCOjMNF/w2cvBiY8e.iXaSZ4i.I3bm7pDu&format=json&railOprIsttCd=DG&lnCd=2&stinCd=220&plfNo=2"

for stinCd in range(201, 251):
    url = base_url + str(stinCd) + "&plfNo=2"
    response = requests.get(url)
    parsed_data = json.loads(response.text)
    print(json.dumps(parsed_data, indent=4, ensure_ascii=False))
    print()  # 결과 간 구분을 위한 개행 추가
```
