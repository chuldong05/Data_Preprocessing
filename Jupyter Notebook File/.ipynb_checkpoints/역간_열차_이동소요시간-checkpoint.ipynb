```python
import pandas as pd

# xlsx 파일 경로
file_path = 'C:/Users/dlwhd/Downloads/2023 Hackton/Data_Preprocessing/열차 시간표 노가다.xlsx'

# 모든 시트 불러오기
xl = pd.read_excel(file_path, sheet_name=None, usecols=lambda x: 'Unnamed' not in x)

# 시트별 데이터 추출
dfs = {}  # 시트별 데이터를 담을 딕셔너리
for sheet_name, df in xl.items():
    dfs[sheet_name] = df

# 시트별 데이터 확인
for sheet_name, df in dfs.items():
    print(f"시트명: {sheet_name}")
    print(df)
    print()  # 시트 간 구분을 위해 개행 추가
```
