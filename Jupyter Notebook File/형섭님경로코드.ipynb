```python
def get_statn(statn):
    a = statn['지하철역ID'].to_list()
    b = statn['이전지하철역ID'].to_list()
    c = statn['다음지하철역ID'].to_list()
    d = statn['지하철호선ID'].to_list()
    e = statn['지하철역명'].to_list()
    for i in range(len(a)):
        a[i] = a[i].replace("\uFEFF", "")
        b[i] = b[i].replace("\uFEFF", "")
        c[i] = c[i].replace("\uFEFF", "")
        d[i] = d[i].replace("\uFEFF", "")
        e[i] = e[i].replace("\uFEFF", "")

    statn['지하철역ID'] = a    
    statn['이전지하철역ID'] = b
    statn['다음지하철역ID'] = c
    statn['지하철호선ID'] = d
    statn['지하철역명'] = e

    return statn

statn = get_statn(statn)
```

```python
import networkx as nx

edges = []
edge_labels = {}


for idx, row in df2.iterrows():
    edges.append((row.역명칭,row.환승이후역명, {"weight":row.weight}))
    edge_labels[(row.역명칭,row.환승이후역명)] = row.환승선명

```

```python
edges
```

```python
G = nx.Graph()

for idx, row in df2.iterrows():
    G.add_node(row.역명칭)

G.add_edges_from(edges)
```

```python
G.nodes
```

```python
source = '사당'
target = '서울'


def (source, target):
    p1 = nx.shortest_path(G, source=source, weight="weight")

    p1to6 = nx.shortest_path(G, source=source, target=target, weight="weight")

    length = nx.shortest_path_length(G, source=source, target=target, weight="weight")

    print("All shortest paths from 1: ", p1)
    print("Shortest path from 1 to 6: ", p1to6)
    print("Length of the shortest path: ", length)

    return p1to6

```
